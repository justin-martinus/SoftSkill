```python
import seaborn as sns
import matplotlib.pyplot as plt

def overlayed_kde_plots(dataframes, column_density_names, classifier, x_label, plot_title):
    # Set Seaborn style and color palette
    sns.set_style("whitegrid")
    sns.set_palette("husl")

    # Create a figure and axis for the plot
    plt.figure(figsize=(8, 5))  # Adjust the figure size as needed
    ax = plt.gca()

    combined_df = pd.concat(dataframes)
    titles = combined_df[classifier].unique()

    # Initialize a dictionary to store median values and colors
    median_values = {}
    colors = iter(sns.color_palette())  # Use the default color palette

    # Loop through DataFrames and plot dots for median values
    for df, title in zip(dataframes, titles):
        # Calculate the median for the current DataFrame
        median_value = np.median(df[column_density_names])

        # Get a distinct color for each dataset
        color = next(colors)

        # Add a dot at the median value along the x-axis
        plt.scatter(median_value, 0, color=color, marker='o', label=f'median ({title}): {median_value:.2f}', zorder=2)

    # Loop through DataFrames and plot KDE curves
    for df, title in zip(dataframes, titles):
        sns.kdeplot(data=df[column_density_names], fill=True, ax=ax, label=title, zorder=1)

    # Add labels and title
    plt.xlabel(x_label)
    plt.ylabel("Density")
    plt.title(plot_title)

    # Customize background color
    ax.set_facecolor("#f0f0f0")

    # Add a legend to distinguish between datasets and median values
    plt.legend()

    # Show the plot
    plt.show()
```

```python
overlayed_kde_plots([df_new_ios, df_new_android, df_new_others], "price", "os", "Price", "Smartphone OS Price Distribution")
```
